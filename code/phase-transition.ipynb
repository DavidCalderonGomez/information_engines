```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import ScalarFormatter

# Load the dataset
data = pd.read_csv("phase-transition.txt", delim_whitespace=True, header=None, encoding="utf-16")

# Assign column names for clarity
data.columns = ['SNR', 'gain']

# Generate the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(data['SNR'], data['gain'], color='orange', label='optimal gain')
plt.xscale('log')
# Use ScalarFormatter to display normal numbers
plt.gca().xaxis.set_major_formatter(ScalarFormatter())

# Adding labels, title, and legend
plt.title('optimal gain vs noise')
plt.xlabel('SNR')
plt.ylabel('gain')
plt.legend(loc=4)

# Display the plot
#plt.show()
plt.savefig('phase-transition.pdf', format='pdf')
```
