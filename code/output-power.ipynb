```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import ScalarFormatter

# Load the dataset
data = pd.read_csv("output-power.txt", delim_whitespace=True, header=None, encoding="utf-16")

# Assign column names for clarity
data.columns = ['SNR', 'power']

# Generate the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(data['SNR'], data['power'], color='orange', label=r'output power $ \langle\dot{F}\rangle$')
plt.xscale('log')
# Use ScalarFormatter to display normal numbers
plt.gca().xaxis.set_major_formatter(ScalarFormatter())
plt.axhline(y=0.3, color='blue', linestyle='dashed', linewidth=1, label= r'$\langle \dot{F} \rangle =0.3$')

# Adding labels, title, and legend
plt.title(r'Output power average $\langle \dot{F} \rangle$ vs noise SNR',fontsize=24,pad=15)
plt.xlabel('SNR',fontsize=20)
plt.ylabel(r'$\langle \dot{F} \rangle (k_B T)$',fontsize=20)
plt.legend(fontsize=18)

# Display the plot
#plt.show()
plt.savefig('output-power.pdf', format='pdf')
```
