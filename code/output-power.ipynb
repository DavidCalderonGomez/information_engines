```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import ScalarFormatter

# Load the dataset
data = pd.read_csv("output-power.txt", delim_whitespace=True, header=None, encoding="utf-16")

# Assign column names for clarity
data.columns = ['SNR', 'power']

# Generate the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(data['SNR'], data['power'], color='orange', label=r'output power $\dot{F}$')
plt.xscale('log')
# Use ScalarFormatter to display normal numbers
plt.gca().xaxis.set_major_formatter(ScalarFormatter())

# Adding labels, title, and legend
plt.title('output power vs noise')
plt.xlabel('SNR')
plt.ylabel(r'$\dot{F} (k_B T)$')
plt.legend(loc=4)

# Display the plot
#plt.show()
plt.savefig('output-power.pdf', format='pdf')
```
