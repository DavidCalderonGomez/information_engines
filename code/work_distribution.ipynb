```python
import numpy as np
import matplotlib.pyplot as plt

# Load data (replace 'data_file.dat' with your actual file path)
data = np.loadtxt('work_distribution.dat')

# Plot the normalized histogram
plt.hist(data, bins=20, density=True, alpha=0.7, edgecolor='black')  # Adjust bins as needed
plt.title("Distribution of work W")
plt.xlabel("Data Value")
plt.ylabel("Probability Density")
#plt.grid(axis='y', alpha=0.75)

# Show the plot
plt.show()
```
