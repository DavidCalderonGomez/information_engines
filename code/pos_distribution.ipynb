```python
import numpy as np
import matplotlib.pyplot as plt

# Load data (replace 'pos_distribution.dat' with your actual file path)
data = np.loadtxt('pos_distribution.dat')

# Plot the normalized histogram
plt.hist(data, bins=20, density=True, alpha=0.7, edgecolor='black')  # Adjust bins as needed
plt.title("Distribution of position after 100 sampling times")
plt.xlabel("Data Value")
plt.ylabel("Probability Density")

plt.show()
```
