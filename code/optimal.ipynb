```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
optimal_data = pd.read_csv('optimal.dat', delim_whitespace=True, header=None)

# Assign column names for clarity
optimal_data.columns = ['Gain', 'Work']

# Generate the scatter plot
plt.figure(figsize=(10, 6))
plt.plot(optimal_data['Gain'], optimal_data['Work'], color='blue', label='Work Data')
plt.axhline(0, color='red', linestyle='--', linewidth=0.75, label='W = 0')

# Adding labels, title, and legend
plt.title('Scatter Plot of Work vs Gain')
plt.xlabel('Time')
plt.ylabel('Work')
plt.legend()

# Display the plot
plt.show()
```
