```python
import matplotlib.pyplot as plt
import pandas as pd

data=pd.read_csv('single_trajectory.dat', delim_whitespace=True, header=None)
# Assign column names for clarity
data.columns = ['Time', 'Position', 'Potential']

# Plotting the time series
plt.figure(figsize=(10, 6))
plt.plot(data['Time'], data['Position'], label='Position',color='red', linewidth=0.75)
plt.plot(data['Time'], data['Potential'], label='Potential',color='black', linewidth=0.75)

# Adding labels, title, and legend
plt.title(r'Single trajectory $\alpha=1.5$, SNR=11', fontsize=14)
plt.xlabel(r'Time ts', fontsize=12)
plt.ylabel(r'Position', fontsize=12)
plt.legend(fontsize=10)
# Display the plot
#plt.show()
plt.savefig('single_trajectory.pdf',format='pdf')
```
