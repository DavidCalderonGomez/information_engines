```python
import matplotlib.pyplot as plt
import pandas as pd

# Load data
datas = pd.read_csv("OU-trajectories.txt", delim_whitespace=True, header=None, encoding="utf-16")

# Assign column names
data = datas.iloc[:-1]  # Exclude the last row
work = datas.iloc[-1, 0]  # Extract work value from the last row
columns = ['Time'] + [f'Series {i}' for i in range(1, len(datas.columns))]
data.columns = columns

# Plotting all series
plt.figure(figsize=(10, 6))

for col in data.columns[1:]:  # Skip time column
    plt.plot(data['Time'], data[col], linewidth=0.1, color='orange')

# Adding labels, title, and legend
plt.title(r'Velocity trajectories for the Ornstein Uhlenbeck process', fontsize=24, pad=15)
plt.xlabel(r'Time', fontsize=20)
plt.ylabel(r'Velocity', fontsize=20)
plt.legend(fontsize=18)

# Save or display the plot
plt.savefig('OU-trajectories.pdf', format='pdf')
#plt.show()
```
