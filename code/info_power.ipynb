```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import ScalarFormatter

# Load the first dataset (output power)
data1 = pd.read_csv("output-power.txt", delim_whitespace=True, header=None, encoding="utf-16-le")
data1.columns = ['SNR', 'power']

# Load the second dataset (mutual information)
data2 = pd.read_csv("modified_data.txt", delim_whitespace=True, header=None, encoding="utf-8")
data2.columns = ['SNR', 'mutualinfo']

plt.figure(figsize=(10, 6))

# Plot the first series: output power
plt.scatter(data1['SNR'], data1['power'], color='orange', label=r'Output power $\langle\dot{F}\rangle$')

# Plot the second series: mutual information
plt.scatter(data2['SNR'], data2['mutualinfo'], color='blue', label=r'Information power $\langle \dot{I_c} \rangle$')
# Set both axes to logarithmic scale
plt.xscale('log')
plt.yscale('log')

# Use ScalarFormatter to display normal numbers on the x-axis
plt.gca().xaxis.set_major_formatter(ScalarFormatter())

# Add labels, title, and legend
plt.title(r'Information Power vs. Noise SNR', fontsize=24, pad=15)
plt.xlabel('SNR', fontsize=20)
plt.ylabel(r'Power $(k_BT)/ \tau_r$', fontsize=20)
plt.legend(fontsize=18)

# Save and display the plot
plt.savefig('information_power.pdf', format='pdf')
#plt.show()
```
